### 1. Necesary modules

```python
#Importing all modules required for ARIMA model.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings("ignore")


#importing all modules required for LSTM neural network model.
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

### 2. Dataset in ARIMA model

```python
#Importing daily values of gold future contracts on COMEX market.
data=pd.read_csv('gold.csv')


# Data used in both models will be between the 13.04.2015 and 13.04.2021.
X=data['Open'][::-1].reset_index()
X=X[len(X)-(365*6):].reset_index()['Open']


"""
Dividing the dataset into test and train data. 
The test data set will have data from last 80 days of the imported time series 
and the training data set will hold the rest remaining data from last 6 years.
"""
X_train=list(X[:len(X)-100])
X_test=list(X[len(X)-100:])
```

### 3.ARIMA functions

```python
def arima_plot_func(X_test,predictions,order):
    """
    The puprose of arima_plot_func function is to show the user: 
    -mean squared error
    -order implemented in the ARIMA model
    -plot related with forecasting the test data
    """
    print(f'mse:{mean_squared_error(X_test,predictions)}, order:{order}')
    plt.plot(predictions,label='predicted values')
    plt.plot(X_test,label='real values')
    plt.legend()
    plt.show()

    
def arima_mse_func(X_train,X_test,
                   order,show_all_plots=True):
    """
    The purpose of arima_mse_func function is to fit the ARIMA model
    with the order imputed by the user and to archive the forecasted
    values of the model. If the parameter show_all_plots is True the
    model will show the real values next to predicted ones in a plot
    
    The function itself returns mean squared error of the predicted
    prices
    """
    predictions=[]
    X_train_mse=X_train.copy()
    X_test_mse=X_test.copy()
    for i in range(len(X_test_mse)):
        arima=ARIMA(X_train_mse,order)
        arima_fit=arima.fit()
        prediction=arima_fit.forecast()[0]
        X_train_mse.append(X_test_mse[i])
        predictions.append(float(prediction))
        
    if show_all_plots:
        arima_plot_func(X_test_mse,predictions,order)

    return mean_squared_error(X_test_mse,predictions)


def arima_best_order_func(X_train,X_test,
                          p_value,d_value,q_value,
                          show_all_plots=True):
    """
    The purpose of arima_best_order_func function is to return the 
    best order, imputed in parameters by the user, of the ARIMA 
    model that is predicting gold prices. The best order is chosen
    by comparison mean squared errors (by the arima_mse_func)
    
    The function returns the best order and the best calculated mean
    squared error.
    """
    best_mse=10000
    best_order=None
    for p in p_value:
        for d in d_value:
            for q in q_value:
                order=(p,d,q)
                try:
                    mse = arima_mse_func(X_train,X_test,
                                   order,
                                   show_all_plots)
                    if best_mse>mse:
                        best_mse=mse
                        best_order=order
                except:
                    continue
    return best_mse,best_order


def arima_predicting_values_func(X_train,X_test,
                                 p_value,d_value,q_value,
                                 days=1,show_all_plots=True):
    """
    The purpose of arima_predicting_values_func function is to predict
    the future values of the gold future contracts with the ARIMA model
    with the best order (by the arima_best_order_func).

    The function returns predicted values, the best calculated mean squared
    error and the best order.
    """
    #Training data for the best order. "bo" in X_train_bo means "best order".
    X_train_bo=[price for price in X]
    predictions=[]
    the_best_mse, the_best_order = arima_best_order_func(X_train,X_test,
                                                         p_value,d_value,q_value,
                                                         show_all_plots)
    for i in range(0,iledni):
        arima=ARIMA(X_train_bo,the_best_order)
        arima_fit=arima.fit()
        prediction=float(arima_fit.forecast()[0])
        predictions.append(prediction)
        X_train_bo.append(prediction)
    return predictions, the_best_mse, the_best_order
```

### 5. Dataset in LSTM model

```python
#Standarization of the data that will be used in neural network.
scaler=MinMaxScaler()
standarize_data=scaler.fit_transform(np.array(X).reshape(-1,1))
standarize_training_data=scaler.fit_transform(np.array(X_train).reshape(-1,1))


#how many days will be used to predict the next future value
pred_days=80


#creating training data that will be used to fit the neural network
mlX_train=[]
mly_train=[]
for i in range(pred_days,len(standarize_training_data)):
    mlX_train.append(standarize_training_data[i-pred_days:i])
    mly_train.append(standarize_training_data[i])


#creating test data that will be used to see how accurate the predictions will be.
mlX_test=[]
mly_test=[]
for i in range(len(standarize_training_data),len(standarize_data)):
    mlX_test.append(standarize_data[i-pred_days:i])
    mly_test.append(standarize_data[i])

    
#changing the form of train and test data to arrays to properly input them in the neural network
mlX_train=np.array(mlX_train)
mly_train=np.array(mly_train)
mlX_test=np.array(mlX_test)
mly_test=np.array(mly_test)

#early stopping necessary for the function that will fit the model
early_stoping=EarlyStopping(monitor='loss',patience=5,restore_best_weights=True)
```

### 6. LSTM neural network functions

```python
def lstm_building_func(layers_units,
                       n_features,
                       dropout,
                       dropout_rate,
                       input_optimizer):
    """
    The purpose of the lstm_builiding_func function is to create the neural
    network with parameters imputed by the user. 
    
    The function returns created neural network. 
    """
    model=Sequential()
    
    for numbers,layer_units in enumerate(layers_units):
        
        if numbers==0:
            model.add(LSTM(units=layer_units,return_sequences=True,input_shape=(n_features,1)))
            if dropout:
                model.add(Dropout(dropout_rate))
                 
        if numbers in range(0,len(layers_units)-1):
                model.add(LSTM(units=layer_units,return_sequences=True))
                if dropout:
                    model.add(Dropout(dropout_rate))
                    
        else:
            model.add(LSTM(units=layer_units))
            if dropout:
                model.add(Dropout(dropout_rate))
                     
    model.add(Dense(1,))
    model.compile(optimizer=input_optimizer,loss='mean_squared_error')
    return model


def lstm_best_model(X_train,y_train,X_test,y_test,
                  models,epochs, batch_size,
                  show_all_plots=True,show_best_plot=True):
    """
    The purpose of lstm_best_model is to predict the future values of the 
    best model in the list of models imputed by a user in parameters of the
    function. If the user wants she/he can visualize all or/and the best 
    predictions of models imputed.
    
    The function returns the best mean squared error and the best LSTM model 
    for predictions.
    """
    best_mse=10000
    best_model=None
    
    for i,model in enumerate(models):
        model.fit(X_train,y_train,epochs=epochs,batch_size=batch_size,callbacks=[early_stoping])
        predicted_values=scaler.inverse_transform(model.predict(X_test))
        mse = mean_squared_error(predicted_values,scaler.inverse_transform(y_test))
        
        if show_all_plots:
            print(f'{i+1} model mse:{mse}')
            plt.plot(predicted_values,label='predicted values')
            plt.plot(scaler.inverse_transform(y_test),label='real values')
            plt.show()
            
        if mse<best_mse:
            best_mse=mse
            best_model=model
            
    if show_best_plot:
        plt.title('best prediction visualization')
        plt.plot(scaler.inverse_transform(best_model.predict(X_test)),label='predicted values')
        plt.plot(scaler.inverse_transform(y_test),label='real values')
        plt.show()
        
    return best_mse     
```
